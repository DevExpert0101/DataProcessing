## Load 'B2B 10080 7182022'

```python
data2 = pd.read_excel('./data/B2B 10080 7182022.xlsx', 'Sheet1')
```

```python
data2['1'] = 'B2B 10080 7182022'
print(data2.head())
```

```python
data2 = data2.rename(columns={'Inventory Location': 'Location'})
data2
```

```python
data2.columns = data2.columns.str.lower()
data2 = data2.reindex(sorted(data2.columns), axis=1)
data2
```

## Combine data1 & data2

```python
for col in default_data.columns:
    if col not in list(data2.columns):
        data2[col] = ""
for col in data2.columns:
    if col not in list(data1.columns):
        data1[col] = ""

data1 = data1.reindex(sorted(data1.columns), axis=1)
data2 = data2.reindex(sorted(data2.columns), axis=1)
```

```python
data1
```

```python
data2
```

```python
default_data = [data1, data2]
default_data = pd.concat(default_data)
default_data
```

```python
default_data.to_excel('output.xlsx')
```
